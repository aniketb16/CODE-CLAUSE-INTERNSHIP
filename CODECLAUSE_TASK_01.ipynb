```python
# @title LOADING LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
# @title LOADING DATASET
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)  #Creating dataframe.
df['target']=iris.target
df.head()  # viewing first 5 rows.
```

```python
# @title PERFORMING EXPLORATARY DATA ANAYLSIS
#  checking dataset rows and columns.
df.shape
#  there are 150 rows and 5 columns.
```

```python
# checking for data types and null values.
df.info()  #  there are no null values,  and we have all numeric data types.
```

```python
df.columns
```

```python
# performing basic statistics
df.describe() # seeing the maximum value and 75% and 25 % we can observe no null values.
```

```python
df.isnull().sum()
```

```python
# visualizing data is crucial to understand realtion between features more accurately.
sns.pairplot(df, hue='target')
 #pair plot is a matrix of graphs that enables the visualization of the relationship between each pair of variables in a dataset.
plt.tight_layout()
# it shows relation between features.
```

```python
plt.figure(figsize=(25,10))    # plotting boxplot to check for outliers present in our data, as we can see sepalwidth have few outliers.

plt.subplot(2,2,1)
sns.boxplot(df['sepal length (cm)'],orient='h')
plt.title('Sepal length ')

plt.subplot(2,2,2)
sns.boxplot(df['sepal width (cm)'],orient='h')
plt.title('SepalWidth')

plt.subplot(2,2,3)
sns.boxplot(df['petal length (cm)'],orient='h',color='orange')
plt.title('PetalLength')

plt.subplot(2,2,4)
sns.boxplot(df['petal width (cm)'],orient='h',color='orange')
plt.title('Petal Width')
plt.tight_layout()
```

```python
#Heatmaps are use to show relationships between two variables, one plotted on each axis.
# values close to 1 shows strong relationship between variables.
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.title('correlation matrix')
plt.tight_layout()
# we can see strong relation between few columns.
```

```python
# countplot helps in visualizing counts of each type.
sns.countplot(x=df['target'], color='orange',edgecolor='black')
# we can see that each type has 50 count.
```

```python
# pie chart of each type, pie chart helps in better visualiation of distribution
plt.figure(figsize=(10,10))
plt.pie(df['target'].value_counts(), labels=df['target'].unique(),autopct='%.2f%%')
plt.title('pie chart of each type')
```

```python
# histogram shows the frequency distribution of the values of the numerical variables.
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.hist(df['petal length (cm)'],bins=10,color='orange',edgecolor='black')
plt.title('histogram of petal length')

plt.subplot(2,2,2,)
plt.hist(df['petal width (cm)'],bins=10,color='orange',edgecolor='black')
plt.title('histogram of petal width')

plt.subplot(2,2,3)
plt.hist(df['sepal length (cm)'],bins=10,color='orange',edgecolor='black')
plt.title('histogram of sepal length')

plt.subplot(2,2,4)
plt.hist(df['sepal width (cm)'],bins=10,color='orange',edgecolor='black')
plt.title('histogram of sepal width')
plt.tight_layout()
```

```python
#A violin plot is similar to box plot and display the same overall statistics of the data,but it also displays the distribution shape for that data.
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.violinplot(data=df, y=df['petal length (cm)'],x=df['target'],color='pink',edgecolor='black')
plt.title('violinplot of petal length')

plt.subplot(2,2,2)
sns.violinplot(data=df,y=df['petal width (cm)'],x=df['target'],color='pink',edgecolor='black')
plt.title('violinplot of petal width')

plt.subplot(2,2,3)
sns.violinplot(data=df,y=df['sepal length (cm)'],x=df['target'],color='green',edgecolor='black')
plt.title('violinplot of sepal length')

plt.subplot(2,2,4)
sns.violinplot(data=df,y=df['sepal width (cm)'],x=df['target'],color='green',edgecolor='black')
plt.title('violinplot of sepal width')
```
